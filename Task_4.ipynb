```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_csv('/content/USvideos.csv')
```

```python
columns_to_remove = ['thumbnail_link','description']
df = df.drop(columns = columns_to_remove)
df.info()
```

```python
from datetime import datetime
```

```python
import datetime
```

```python
df["trending_date"] = df ["trending_date"].apply(lambda x : datetime.datetime.strptime(x, '%y.%d.%m'))
df.head(3)
```

```python
df['publish_time'] = pd.to_datetime(df['publish_time'])
df.head(2)
```

```python
df['publish_month'] = df['publish_time'].dt.month
df['publish_day'] = df['publish_time'].dt.day
df['publish_hour'] = df['publish_time'].dt.hour
df.head(2)
```

```python
print (sorted(df["category_id"].unique()))
[1, 2, 10, 15, 17, 19, 20, 22, 23, 24, 25, 26, 27, 28, 29, 30, 43]
```

```python
df['year'] = df['publish_time'].dt.year
yearly_counts = df.groupby('year')['video_id'].count()

#create a bar chart
yearly_counts.plot(kind='bar', xlabel='year', ylabel= 'Total Publish Count', title= 'Total publish Videos Per Year')

#show the count
plt.show()
```

```python
#Group by year and sum the views for each year
yearly_views = df.groupby('year')['views'].sum()

#create a bar chart
yearly_views.plot(kind='bar', xlabel='year', ylabel= 'Total Views', title= 'Total Views Per Year')
plt.xticks(rotation=0)
plt.tight_layout() #check visulations is not overlap

#show the count
plt.show()
```

```python
category_mapping = {
    1: "Film & Animation",          # Example: "Top 10 Best Movies of 2023"
    2: "Autos & Vehicles",          # Example: "Reviewing the New Tesla Model S"
    3: "Music",                     # Example: "Latest Music Hits of the Week"
    4: "Pets & Animals",            # Example: "How to Train Your Dog"
    5: "Sports",                    # Example: "Top Football Goals of the Year"
    6: "Short Movies",              # Example: "Creating a Short Film: Behind the Scenes"
    7: "Travel & Events",           # Example: "Exploring the Wonders of the Amazon Rainforest"
    8: "Gaming",                    # Example: "Epic Gaming Moments Compilation"
    9: "Videoblogging",             # Example: "Daily Vlog: A Day in My Life"
    10: "People & Blogs",           # Example: "Family Fun Time: Visiting the Zoo"
    11: "Comedy",                   # Example: "Hilarious Stand-Up Comedy"
    12: "Entertainment",            # General Entertainment content
    13: "News & Politics",          # Example: "Breaking News: Major Earthquake in City"
    14: "How-to & Style",           # Example: "DIY Home Decor Ideas"
    15: "Education",                # Example: "Educational Video: Learning Algebra"
    16: "Science & Technology",     # Example: "The Science Behind Black Holes"
    17: "Nonprofits & Activism",    # Content related to activism or nonprofit work
    18: "Movies",                   # Example: "Action-Packed Movie Trailers"
    19: "Anime/Animation",          # Example: "Anime Review: Attack on Titan"
    20: "Action/Adventure",         # Action content or adventure videos
    21: "Classics",                 # Example: "Classic Hollywood Movies: A Retrospective"
    22: "Comedy (again due to redundancy)", # Example: "Funny Animal Videos Compilation"
    23: "Documentary",              # Example: "Understanding Climate Change"
    24: "Drama",                    # Dramatic content, could be narrative-driven
    25: "Family",                   # Family-oriented content
    26: "Foreign",                  # Content related to foreign films or cultures
    27: "Horror",                   # Example: "Thrilling Horror Movie Clips"
    28: "Thriller"                  # Suspense or thriller content
}


# Add the 'category_name' column using the mapping
df['category_name'] = df['category_id'].map(category_mapping)
```

```python
#Group the data by 'category_name' and calculate the sum of 'views' in each category
category_views = df.groupby('category_name')['views'].sum().reset_index()

#short the categories by vuews in desending order
top_categories = category_views.sort_values(by='views', ascending = False).head(5)

#create a bar plot fo visualize the top 5 categories
plt.bar(top_categories['category_name'], top_categories['views'])
plt.xlabel('Category Name', fontsize=12)
plt.ylabel('Total Views', fontsize=12)
plt.title('Top 5 Categories Views', fontsize=12)
plt.tight_layout() #check visulations is not overlap

#show the count
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='category_name', data=df, order=df['category_name'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Video Count by category')
plt.show()
```

```python
df['publish_time'] = pd.to_datetime(df['publish_time'])
df['publish_date'] = df['publish_time'].dt.date
video_count_by_date = df.groupby('publish_date').size()
plt.figure(figsize=(12, 6))
sns.lineplot(data=video_count_by_date)
plt.title('Video Published Over Time')
plt.xlabel('Published Date')
plt.ylabel('Nimber of Videos')
plt.xticks(rotation=45)
plt.show()
```

```python
#scatter plot between 'views' and 'likes'
sns.scatterplot(data=df, x= 'views', y= 'likes')
plt.title('Views vd likes')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()
```
